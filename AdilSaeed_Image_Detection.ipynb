```python
# Function to run inference on a single image
def infer_image(image_path, output_path=None):
    # Read the image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Could not load image at {image_path}")
        return

    # Perform inference using your trained YOLO model
    results = model(img)

    # Draw bounding boxes and labels on the image
    annotated_img = results[0].plot()  # YOLO v8+ plot method

    # Show the image in notebook
    cv2_imshow(annotated_img)

    # Save the output image if path is provided
    if output_path:
        cv2.imwrite(output_path, annotated_img)
        print(f"Output image saved at {output_path}")

# Example usage for your new project
image_path = "/content/Adil/test/images/CV_Project_Detection_Images_pdf_page_21_png.rf.25b67f6b1ab57fad3c81fd7e7f854131.jpg"
output_path = "/content/Adil/test/HumansAndAnimals_Detection256/output_image.jpg"

infer_image(image_path, output_path)
```

```python
import cv2

def infer_video(video_path, output_path="/content/output_video_10000.mp4"):
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Could not open video at {video_path}")
        return

    # Get video properties
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # Process video frame by frame
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Perform inference on the frame
        results = model1(frame)

        # Get the annotated frame (with bounding boxes)
        annotated_frame = results[0].plot()

        # Access detection results (boxes, scores, and classes)
        detections = results[0].boxes  # Assuming YOLO-like model output
        for box in detections:
            # Extract coordinates, confidence, and class ID
            x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)  # Bounding box coordinates
            conf = box.conf[0].cpu().numpy()  # Confidence score
            cls = int(box.cls[0].cpu().numpy())  # Class ID
            label = f"{model1.names[cls]} {conf:.2f}"  # Class name and confidence score

            # Define text properties
            font = cv2.FONT_HERSHEY_SIMPLEX
            font_scale = 0.5
            color = (255, 255, 255)  # White text
            thickness = 1
            text_size = cv2.getTextSize(label, font, font_scale, thickness)[0]
            text_x = x1
            text_y = y1 - 10 if y1 - 10 > 10 else y1 + 10  # Place text above or below box

            # Draw text background for better visibility
            cv2.rectangle(
                annotated_frame,
                (text_x, text_y - text_size[1]),
                (text_x + text_size[0], text_y),
                (0, 0, 0),  # Black background
                cv2.FILLED,
            )

            # Draw text on the frame
            cv2.putText(
                annotated_frame,
                label,
                (text_x, text_y),
                font,
                font_scale,
                color,
                thickness,
                cv2.LINE_AA,
            )

        # Write the frame to the output video
        out.write(annotated_frame)

    # Release resources
    cap.release()
    out.release()
    cv2.destroyAllWindows()
    print(f"Output video saved at {output_path}")
```

```python
# Open default webcam (index 0)
import cv2
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Cannot open webcam")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Run YOLO inference on the frame (returns results)
    results = model(frame)

    # results[0].plot() returns an image with detections drawn on it
    annotated_frame = results[0].plot()

    # Show the frame with detections
    cv2.imshow('YOLOv8 Live Webcam', annotated_frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
